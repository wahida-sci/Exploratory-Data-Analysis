```python
mean_normalized = mydata['normalized-losses'].astype('float').mean(axis=0)

mean_normalized
```

```python
mydata.head()
```

```python
mydata['normalized-losses'].replace(np.nan, mean_normalized, inplace=True)

mydata.head()
```

```python
mydata['num-of-doors'].value_counts()
```

```python
mydata['num-of-doors'].value_counts().idxmax()
```

```python
mydata['num-of-doors'].replace(np.nan, 'four', inplace=True)

mydata.head()
```

```python
mean_bore = mydata['bore'].astype('float').mean(axis=0)

mean_stroke = mydata['stroke'].astype('float').mean(axis=0)

mean_horsepower = mydata['horsepower'].astype('float').mean(axis=0)

mean_peak = mydata['peak-rpm'].astype('float').mean(axis=0)

print (mean_bore)
print (mean_stroke)
print (mean_horsepower)
print (mean_peak)
```

```python
mydata['bore'].replace(np.nan, mean_bore, inplace=True)

mydata['stroke'].replace(np.nan, mean_stroke, inplace=True)

mydata['horsepower'].replace(np.nan, mean_horsepower, inplace=True)

mydata['peak-rpm'].replace(np.nan, mean_peak, inplace=True)

```

```python
mydata.dropna(subset = ['price'], axis=0, inplace=True)
```

```python
after_missing_value = mydata.isnull()

after_missing_value
```

```python
for column in after_missing_value.columns.values.tolist():
    print (column)
    print (after_missing_value[column].value_counts())
    print ('')
```
